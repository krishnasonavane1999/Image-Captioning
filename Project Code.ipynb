```python
# Read text captions
def readTextFile(path):
    with open(path) as f:
        captions = f.read()
    return captions
```

```python
captions = readTextFile("captions.txt")
len(captions.split("\n"))
captions = captions.split("\n")
len(captions)
captions[-1].split(',')
```

```python
# dictionary to map image and corresponding captions 
```

```python
descriptions = {} 

for x in captions:
    f = x.split(',')
    img_name = f[0].split('.')[0]
    second = f[1]
    
    if descriptions.get(img_name) is None:
        descriptions[img_name] = []
    
    descriptions[img_name].append(second)
```

```python
# Data Cleaning
```

```python
def clean_text(sentence):
    sentence = sentence.lower()
    sentence = re.sub("[^a-z]+"," ",sentence) # regular expression
    sentence = sentence.split()
    
    sentence = [s for s in sentence if len(s)>1]
    sentence = " ".join(sentence)
    return sentence
    
    
```

```python
clean_text("Hello, my name is krishna & 5 21")
```

```python
# Now clean all the captions
for key, caption_list in descriptions.items():
    for i in range(len(caption_list)):
        caption_list[i] = clean_text(caption_list[i]) # clean each caption
```

```python
print(type(descriptions))
print(len(descriptions))
```

```python
# creating vocabulary of unique words
vocab = set()

for key in descriptions.keys():
    sentence = descriptions[key]
    [vocab.update(sentence.split()) for sentence in descriptions[key]]
    
print(len(vocab)) # unique words
```

```python
# Toatal no of words across all the sentences
total_words = []
for key in descriptions.keys():
    [total_words.append(i) for des in descriptions[key] for i in des.split()]
    
print(len(total_words)) # total words available
print(total_words[:10]) # total words available
```

```python
# filter the words from the vocab according to the certain frequency
```

```python
import collections

counter = collections.Counter(total_words)
freq_cnt = dict(counter)
print(len(freq_cnt.keys()))
```

```python
# now sort the dictionary according to the frequency

sorted_freq_cnt = sorted(freq_cnt.items(), reverse = True, key = lambda x : x[1])

# filter
threshold = 10
sorted_freq_cnt = [x for x in sorted_freq_cnt if x[1] > threshold]
total_words = [x[0] for x in sorted_freq_cnt]

print(len(total_words)) # final vocab size

```
